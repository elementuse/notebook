```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# 读取图片
filename = 'cat.jpg';
ori_img = np.array(Image.open(filename).convert('L'))
```

```python
# SVD分解
u, sigma, vt = np.linalg.svd(ori_img)
```

```python
# 将奇异值向量的前k个奇异值转为对角矩阵
k = 10
sig = np.eye(k) * sigma[:k]

# 生成压缩图片
img1 = np.dot(np.dot(u[:,:k], sig), vt[:k,:])
```

```python
# 重复生成多个图片测试
k = 20
sig = np.eye(k) * sigma[:k]
img2 = np.dot(np.dot(u[:,:k], sig), vt[:k,:])

k = 30
sig = np.eye(k) * sigma[:k]
img3 = np.dot(np.dot(u[:,:k], sig), vt[:k,:])
```

```python
# 显示图片
fig, ax = plt.subplots(2, 2)
ax[0,0].imshow(img1, cmap='gray')
ax[0,0].set_title('10')
ax[0,1].imshow(img2, cmap='gray')
ax[0,1].set_title('20')
ax[1,0].imshow(img3, cmap='gray')
ax[1,0].set_title('30')
ax[1,1].imshow(ori_img, cmap='gray')
ax[1,1].set_title('origin')
plt.show()
```
