```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# 读取图片
filename = 'dog.jpg';
ori_img = np.array(Image.open(filename))
```

```python
def rebuild_img(u, sigma, v, p): #p表示奇异值的百分比
    m = len(u)
    n = len(v)
    a = np.zeros((m, n))
    
    count = (int)(sum(sigma))
    curSum = 0
    k = 0
    while curSum <= count * p:
        uk = u[:, k].reshape(m, 1)
        vk = v[k].reshape(1, n)
        a += sigma[k] * np.dot(uk, vk)
        curSum += sigma[k]
        k += 1
    a[a < 0] = 0
    a[a > 255] = 255
    #按照最近距离取整数，并设置参数类型为uint8
    return np.rint(a).astype("uint8")
```

```python
p = 0.7

u, sigma, v = np.linalg.svd(ori_img[:, :, 0])
R = rebuild_img(u, sigma, v, p)
        
u, sigma, v = np.linalg.svd(ori_img[:, :, 1])
G = rebuild_img(u, sigma, v, p)
        
u, sigma, v = np.linalg.svd(ori_img[:, :, 2])
B = rebuild_img(u, sigma, v, p)
        
I = np.stack((R, G, B), 2)
plt.imshow(I)
```
